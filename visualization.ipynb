```python
def display_load_and_temperature(equipment_id):
    df = alldf[alldf['equipmentId'] == equipment_id]
    
    # Convertir dateTime en datetime si ce n'est pas déjà fait
    df['dateTime'] = pd.to_datetime(df['dateTime'])
    
    # Trier les données par dateTime
    df = df.sort_values(by='dateTime')

    # Sélectionner les colonnes pertinentes
    df = df[['dateTime', 'load', 'hotspotTemperature']]
    
    # Créer la figure
    plt.figure(figsize=(15, 5))
    
    # Créer deux axes
    ax1 = plt.gca()
    ax2 = ax1.twinx()
    
    # Tracer la charge (load) sur l'axe de gauche
    ax1.plot(df['dateTime'], df['load'], 'b-', label='Load')
    ax1.set_ylabel('Load', color='b')
    ax1.set_xlabel('DateTime')
    
    # Tracer la température du hotspot sur l'axe de droite
    ax2.plot(df['dateTime'], df['hotspotTemperature'], 'r-', label='Hotspot Temperature')
    ax2.set_ylabel('Temperature (°C)', color='r')
    
    # Ajouter des légendes
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
    
    # Ajuster l'affichage des dates
    plt.xticks(rotation=45)
    
    plt.title(f"Load and Hotspot Temperature for Equipment {equipment_id}")
    plt.show()

```

```python
import warnings
import re
import matplotlib.pyplot as plt
import pandas as pd
from itertools import cycle

def plot_by_units(
    equipment_id,
    columns,
    columns_map,      # dict: column -> {'units': ..., 'description': ...}
    df_source=None,   # defaults to global alldf if None
    time_window=None  # e.g., ['2023-05','2023-09'] or ['2023-05-10','2023-06-02']
):
    """
    Plot time series for a given equipment_id, grouped by units:
      - Columns with unit '-' are skipped (warned).
      - If >2 distinct units among requested columns, warn and keep the first two encountered.
      - Legends show 'column [unit]'.
      - First unit group -> left axis; second unit group -> right axis.
      - Distinct colors across all series (no repetition).
      - Optional time_window filtering: inclusive between start and end.
        * 'YYYY-MM' expands to the whole month.
        * 'YYYY-MM-DD' uses that exact day.
    """
    if df_source is None:
        if 'alldf' not in globals():
            raise ValueError("df_source is None and global 'alldf' not found.")
        df_source = alldf

    # --- Filter rows for this equipment ---
    df = df_source[df_source['equipmentId'] == equipment_id].copy()
    if df.empty:
        warnings.warn(f"No data for equipmentId={equipment_id}.")
        return

    # --- Ensure datetime and sort ---
    df['dateTime'] = pd.to_datetime(df['dateTime'], errors='coerce')
    df = df.dropna(subset=['dateTime']).sort_values('dateTime')

    # --- Apply time window if provided ---
    def _parse_bound(s, is_end=False):
        ts = pd.to_datetime(s, errors='coerce')
        if ts is pd.NaT:
            raise ValueError(f"Invalid time bound: {s}")
        # If only YYYY-MM provided, expand end to month end
        if isinstance(s, str) and re.fullmatch(r"\d{4}-\d{2}", s):
            if is_end:
                ts = ts + pd.offsets.MonthEnd(1)  # end of that month
        return ts

    if time_window is not None:
        if not (isinstance(time_window, (list, tuple)) and len(time_window) == 2):
            raise ValueError("time_window must be a 2-element list/tuple like ['2023-05','2023-09'].")
        start = _parse_bound(time_window[0], is_end=False)
        end   = _parse_bound(time_window[1], is_end=True)
        df = df[(df['dateTime'] >= start) & (df['dateTime'] <= end)]
        if df.empty:
            warnings.warn(f"No data after applying time_window {time_window}.")
            return

    # --- Resolve units for requested columns; skip missing or non-physical ('-') ---
    resolved = []
    for col in columns:
        if col not in df.columns:
            warnings.warn(f"Column '{col}' not found in dataframe; skipping.")
            continue
        meta = columns_map.get(col, {})
        unit = meta.get('units')
        if unit is None:
            warnings.warn(f"No unit defined for '{col}' in columns_map; skipping.")
            continue
        if unit == '-':
            warnings.warn(f"Unit for '{col}' is '-' (non-physical); skipping.")
            continue
        resolved.append((col, unit))

    if not resolved:
        warnings.warn("No plottable columns after unit checks.")
        return

    # --- Enforce at most 2 units (keep order of first appearance) ---
    units_in_order = []
    for _, u in resolved:
        if u not in units_in_order:
            units_in_order.append(u)
    if len(units_in_order) > 2:
        warnings.warn(
            f"More than two units requested ({units_in_order}). "
            f"Only the first two will be plotted."
        )
        allowed_units = set(units_in_order[:2])
        resolved = [(c, u) for (c, u) in resolved if u in allowed_units]
        units_in_order = units_in_order[:2]

    # --- Group columns by unit (preserving requested order) ---
    by_unit = {u: [] for u in units_in_order}
    for c, u in resolved:
        by_unit[u].append(c)

    # --- Prepare distinct colors across ALL series ---
    # Use matplotlib's default color cycle and extend if needed
    base_colors = plt.rcParams['axes.prop_cycle'].by_key().get('color', [
        # safe fallback palette
        "#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd",
        "#8c564b","#e377c2","#7f7f7f","#bcbd22","#17becf"
    ])
    needed = sum(len(cols) for cols in by_unit.values())
    color_list = (base_colors * ((needed // len(base_colors)) + 1))[:needed]
    color_iter = iter(color_list)

    # --- Plot ---
    fig, ax_left = plt.subplots(figsize=(15, 5))
    ax_right = None

    # Left axis: first unit
    left_unit = units_in_order[0]
    for col in by_unit[left_unit]:
        ax_left.plot(df['dateTime'], df[col], label=f"{col} [{left_unit}]", color=next(color_iter))
    ax_left.set_xlabel("DateTime")
    ax_left.set_ylabel(left_unit)

    # Right axis if a second unit exists
    if len(units_in_order) == 2:
        right_unit = units_in_order[1]
        ax_right = ax_left.twinx()
        for col in by_unit[right_unit]:
            ax_right.plot(df['dateTime'], df[col], label=f"{col} [{right_unit}]", color=next(color_iter))
        ax_right.set_ylabel(right_unit)

    # Combined legend
    lines_l, labels_l = ax_left.get_legend_handles_labels()
    lines = lines_l
    labels = labels_l
    if ax_right is not None:
        lines_r, labels_r = ax_right.get_legend_handles_labels()
        lines += lines_r
        labels += labels_r
    ax_left.legend(lines, labels, loc="upper right")

    plt.xticks(rotation=45)
    requested_cols = ", ".join(columns)
    title = f"Equipment {equipment_id} – {requested_cols}"
    if time_window is not None:
        title += f" | Window: {time_window[0]} → {time_window[1]}"
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
columns = ['load', 'hotspotTemperature', 'temperature']
time_window = ['2023-06-25', '2023-06-30']
plot_by_units(50282127, columns, columns_map, time_window=time_window)
```
